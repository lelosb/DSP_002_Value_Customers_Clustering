### Gross Revenue

```python
#Gross Revenue(Faturamento)
df3_purchases['gross_revenue'] = df3_purchases['quantity'] * df3_purchases['unit_price']
#Monetary (quanto a pessoa já gastou com a gente)
df_monetary = df3_purchases[['customer_id','gross_revenue']].groupby('customer_id').sum().reset_index()
df_ref = pd.merge(df_ref, df_monetary, on ='customer_id', how='left')
#Concatena o df_ref e o df_monetary pela coluna customer id tendo a coluna da esquerda como referencia
df_ref.isna().sum()
```

### Recency - Day from last purchase

Recencia - Em uma situação normal, faríamos a conta utilizando uma função para a data do dia, de forma que o modelo fosse se atualizando todos os dias. No nosso caso, temos uma base fechada em um período fechado, então é mais lógico usar a última compra do data frame como parâmetro para as outras

```python
df_recency = df3_purchases[['customer_id', 'invoice_date']].groupby('customer_id').max().reset_index()
#retorna a data máxima por cliente
df_recency['recency_days'] = (df3_purchases['invoice_date'].max() - df_recency['invoice_date']).dt.days
df_recency = df_recency[['customer_id', 'recency_days']].copy() #pra fazer uma cópia fisica
df_ref = pd.merge(df_ref,df_recency, on ='customer_id', how='left') 
# una as tabelas ref e recency baseado no customer_id com a tabela da esquerda como referência
df_ref.isna().sum()
```

###  Quantity of Purchases

```python
#number of purchases per customer
df_purchase = df3_purchases[['customer_id', 'invoice_no']].drop_duplicates().groupby('customer_id').count().reset_index().rename(columns = {'invoice_no': 'qtd_invoices'})
#o drop é pra tirar os invoices repetidos, o count é pra somar a quantidade de invoices de cada cliente
#drop vem como padrão 'keep first', dropa os duplicados de todas as colunas, exceto da primeira
df_ref = pd.merge (df_ref, df_purchase, on='customer_id', how='left')
df_ref.isna().sum()
```

###  Total quantity of Items purchased

```python
#total number of items per customer
df_quantity = df3_purchases[['customer_id', 'quantity']].groupby('customer_id').sum().reset_index().rename(columns={'quantity': 'qtd_items'})
#drop vem como padrão 'keep first', dropa os duplicados de todas as colunas, exceto da primeira
df_ref = pd.merge(df_ref, df_quantity, on='customer_id', how='left')
df_ref.isna().sum()
# Aqui cabe uma observação. Cometemos o erro crasso de dropar os duplicados na quantidade, o que não faz sentido nenhum. Além disso,
# contamos ao inves de somar. Esse erro só foi pego quando olhamos os outliers
```

###  Unique Basket Size - Quantity of distinct products purchased per invoice

```python
df_ubasket = df3_purchases[['customer_id', 'invoice_no', 'stock_code']].groupby('customer_id').agg ( n_purchase = ('invoice_no', 'nunique') , n_products = ('stock_code', 'nunique')).reset_index()

df_ubasket['avg_unique_basket_size'] = df_ubasket['n_products'] / df_ubasket['n_purchase']

df_ref = pd.merge (df_ref, df_ubasket[['customer_id', 'avg_unique_basket_size']], on = 'customer_id', how ='left')

df_ref.isna().sum()
```

### Avg Ticket Value

```python
#Avg Ticket
df_avg_ticket = df3_purchases[['customer_id','gross_revenue']].groupby('customer_id').mean().reset_index().rename( columns = {'gross_revenue': 'avg_ticket'})
df_ref = pd.merge(df_ref, df_avg_ticket, how ='left' , on='customer_id')
df_ref.isna().sum()

#tem um cliente que comprou e devolveu tudo, isso foi pego na analise do pandaprofiling. Tem que decidir o que vai ser feito com as devoluções
```

### Frequency Purchase

```python
#This feature does not count the total time of the dataset, just the time and number of invoices between the first and last
# purchase for each customer
# .agg function
df_aux = (df3_purchases[['customer_id', 'invoice_no',
               'invoice_date']].drop_duplicates().groupby('customer_id').agg(
                   max_=('invoice_date', 'max'), #nova coluna, sobre qual coluna, qual operação
                   min_=('invoice_date', 'min'),
                   days_=('invoice_date', lambda x:
                          ((x.max() - x.min()).days) + 1),
                   buy_=('invoice_no', 'count'))).reset_index()
# the '1' sum is because the minimum frequency is 1. If the customer just bought one time, the date will be the same and the difference
#will be 0

#frequency
df_aux['frequency'] = df_aux[['buy_','days_']].apply(lambda x: x['buy_'] / x['days_']if x['days_'] != 0 else 0,axis=1)
#merge
df_ref = pd.merge(df_ref,df_aux[['customer_id', 'frequency']],on='customer_id', how='left')
df_ref.isna().sum()
```

```python
df_ref['frequency'].max()
```

## Space study

Antes de fazer o estudo dos espaços é necessário ajustar a escala dos dados. Podemos utilizar o standard scaller ou o min/max scaler ou até o robust scaller

```python
df43 = df4.drop(columns = ['customer_id'], axis=1).copy() #não vamos reescalar o costumer_id
```

```python
mm = pp.MinMaxScaler()

# Cada coluna é fitada separadamente
df43['gross_revenue']          = mm.fit_transform (df43[['gross_revenue']]) # tem que passar como lista
df43['recency_days']           = mm.fit_transform (df43[['recency_days']])
df43['qtd_invoices']           = mm.fit_transform (df43[['qtd_invoices']])
df43['qtd_items']              = mm.fit_transform (df43[['qtd_items']])
df43['avg_ticket']             = mm.fit_transform (df43[['avg_ticket']])
df43['frequency']              = mm.fit_transform (df43[['frequency']])
df43['qtd_returns']            = mm.fit_transform (df43[['qtd_returns']])
df43['avg_unique_basket_size'] = mm.fit_transform (df43[['avg_unique_basket_size']])
df43['qtd_returns']            = mm.fit_transform (df43[['qtd_returns']])
df43['avg_invoice_value']      = mm.fit_transform (df43[['avg_invoice_value']])

#Dados reescalados, vamos tentar encontrar espaços melhores
X = df43.copy()
```

### PCA

```python
pca = dd.PCA (n_components = X.shape[1]) 
#passa como número de componente a qtd de colunas do dataframe. 
#Cria um espaço de dimensão igual ao número de features

principal_components = pca.fit_transform( X )

#plot explained variable
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_ratio_ , color='black')
```

O tamanho das barras representa a variação das variáveis. Nesse caso, quer dizer que as variáveis 0 e 1 são as que possuem maior variância, por isso seriam as ideais para gerar o espaço

```python
df_pca = pd.DataFrame (principal_components)
df_pca.head()
```

```python
sns.scatterplot(x=0, y=1, data= df_pca) # 0 e 1 por causa da maior variação que foi descoberta no item anterior
# as colunas não tem nome, logo os números são os índices
```

### Tree-Based Embedding

Truque: vamos fazer como se fosse um problema com resposta final. Vamos usar o gross revenue como variável resposta e gerar as arvores

```python
#training dataset
X = df4.drop (columns = ['customer_id', 'gross_revenue'], axis = 1) #variáveis indepentes
Y = df4['gross_revenue']                                           #variavel dependente (saída)

#model definition
rf_model = en.RandomForestRegressor ( n_estimators = 100 , random_state = 42)

#model training
rf_model.fit (X,Y)
```

```python
#A árvore neste ponto já está treinada. Agora pegamos a arvore treinada e aplicamos os dados para pegar as folhas

df_leaf = pd.DataFrame (rf_model.apply( X )) #aplicando a arvore treinada nos dados de treino e salvando em um dataframe
df_leaf.head()
```

## K-Means

```python
kmeans_list = []
for k in clusters:
    #model definition
    kmeans_model = c.KMeans (n_clusters = k) 
    #model training
    kmeans_model.fit( X )
    #model predict
    labels = kmeans_model.predict(X)
    #model performance
    sil = m.silhouette_score (X, labels, metric = 'euclidean')
    kmeans_list.append(sil)
    
```

## GMM

```python
gmm_list=[]
for k in clusters:
    #model definition
    gmm_model = mx.GaussianMixture (n_components = k) #n_components seria a qtd de gaussianas
    #model training
    gmm_model.fit( X )
    #model predict
    labels = gmm_model.predict(X)
    #model performance
    sil = m.silhouette_score (X, labels, metric = 'euclidean')
    gmm_list.append ( sil )
```

```python
plt.plot (clusters, gmm_list, linestyle = '--', marker='o', color = 'b')
plt.xlabel('K')
plt.ylabel('Silhouette Score')
plt.title ('Silhouette Score X K')
```

```python
eps = 0.1 #raio épsilon, não pode ser muito maior que a grandeza das features
min_samples = 200 # quantidade de pontos da vizinhança
#model definition
dbscan_model = c.DBSCAN (eps=eps, min_samples = min_samples)
#model training and predict
labels = dbscan_model.fit_predict ( X ) # labels são os clusters

sil_db = m.silhouette_score (X, labels, metric='euclidean')
print ('Silhouette Score: {}'.format (sil_db))
print ('Number of Clusters: {}'. format (len(unique(labels))))
```

```python
#Como encontrar o melhor k?
#A idéia é plotar e onde o gráfico fizer a curvinha é o valor do raio a ser colocado

neighbors = NearestNeighbors (n_neighbors = min_samples).fit(X) #retorna uma matriz de distâncias dos vizinhos

distances , indices = neighbors.kneighbors (X) #o método K.neighbors retorna dois valores, por isso a vírgula

distances = np.sort (distances, axis = 0) #ordena a matriz de distâncias

distances = distances[:,1] #pega só a primeira linha da matriz, pois a matriz tem a distância de todos pra todos, só primeira linha
#já tem toda a informação que precisamos

plt.plot (distances)
```

```python
# Não tem silhouette no dbscan, o Meigarom fez do zero
fig , ax = plt.subplots (3,2)
fig.set_size_inches(25 , 20)

for k in clusters:
    q, mod = divmod( k , 2 ) #truque pra plotar cada gráfico em uma posição
    ax[q-1,mod].set_xlim ([-0.1, 1 ]) #configuração do eixo x
    
    ax[q-1,mod].set_ylim ([0, len(X) + (k+1)*10]) #configuração do eixo y
    
    #model definition and training
    hc_model = hc.linkage (X, 'ward')
    #model predict
    labels = hc.fcluster (hc_model, k , criterion='maxclust')
    #performance
    ss= m.silhouette_score (X, labels, metric = 'euclidean')
    print ('For K={}. Silhouette Score {}'.format (k, ss))
    samples_silhouette_values = m.silhouette_samples (X, labels) #pegamos os valores dos pontos pra fazer o gráfico de silhoueta
    y_lower = 10 # espaço entre os clusters
    for i in range (k):
        #select clusters
        ith_samples_silhouette_values = samples_silhouette_values[labels==i] #i-ésimo valor de silhouette
        #sort_values
        ith_samples_silhouette_values.sort()
        #size clusters
        size_cluster_i = ith_samples_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i #espaço entre os clusters. O espaço mais o tamanho do cluster
        cmap = cm.get_cmap ('Spectral')
        color = cmap (i/k)
        
        ax[q-1, mod].fill_betweenx(np.arange(y_lower, y_upper), 0, ith_samples_silhouette_values)
        #fill_between pra fazer o gráfico preenchido ao longo de x
        y_lower = y_upper + 10
        
        ax[q-1, mod].set_yticks([])
        ax[q-1, mod].set_xticks([-0.1,0,0.2,0.4,0.6,0.8,1])
        
```

## Within-Cluster Sum of Square (WSS)

    A métrica wss é utilizada para encontrar qual o numero de clusters (dentro da quantidade que escolhemos pra treinar o modelo) é o melhor para resolver o problema, por isso plotamos wss X k.
    Essa métrica soma as distâncias DENTRO dos clusters, não entre os clusters, o que pode ser um problema.

```python
#implementando do modo roots (depois dá pra usar uma solução pronta)
wss = []
for k in clusters:
    #model definition
    kmeans = c.KMeans (init='random', n_clusters=k, n_init=10, max_iter=300, random_state=42)
    #model training
    kmeans.fit(X)
    #validation
    wss.append(kmeans.inertia_) #ws é uma métrica de validação
#plot wss
plt.plot(clusters, wss, linestyle='--', marker='o', color='b'); #não entendi o pq do ponto e vírgula
plt.xlabel('K');
plt.ylabel('Within-Clster Sum of Square');
plt.title('WSS vs. K');
```

```python
#forma mais elegante

kmeans = KElbowVisualizer (c.KMeans(), k=clusters, timings=False)
kmeans.fit(X)
kmeans.show()                         
```
